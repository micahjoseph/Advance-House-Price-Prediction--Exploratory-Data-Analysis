**IDENTIFYING THE MISSING VALUES**

```python
#Make the list of features with missing values
feature_with_na=[feature for feature in df.columns if df[feature].isnull().sum()>1]
#Print the feature and the corresponding % missing values
for feature in feature_with_na:
  print(feature,np.round(df[feature].isnull().mean(),4),'% missing values')
```

**Since there are many missing values we need to find the relationship between missing values and sales price**


For this we will plot a diagram for the relationship:

```python
for feature in feature_with_na:
  data=df.copy()
  #Let's make a variable 1 if the value is Null or else it's 0
  data[feature]=np.where(data[feature].isnull(),1,0)


  #Let's calculate the mean sale's price where the information is missing or  present
  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.title(feature)
  plt.show()
```

Since the relationship between the Nan values and sales price is clearly visible we need to do replace that with meaningfull values  which would be handled in the next section.

**NUMERICAL FEATURES**

```python
#list of numerical features
numerical_features=[feature for feature  in df.columns if df[feature].dtype!='O']

print('number of numerical feature :', len(numerical_features))
```

```python
df[numerical_features].head()
```

Temporal Features (E.g. DateTime features)
From dataset we can extract information like number of years and and number of days.
We can also identify the difference between when the house was built an when it was sold.
The analysis will be performed in feature engineering part.




```python
year_feature= [feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
print(year_feature)
```

Explore the content of year features

```python
for i in df[year_feature]:
  print(i,df[i].unique())
```

Let's analyse the temporal datetime variables

We will try to analyse whether there is a relationship between YrSold and SalesPrice


```python
df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('YearSold')
plt.ylabel('Median House Price')
plt.title('Median House Price')
```

```python
year_feature
```

Here we will compare difference between all year feature with SalePrice

```python
for feature in year_feature:
  if feature!='YrSold':
    data=df.copy()
    data[feature]=data['YrSold']-data[feature]
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()
```

```python
discrete_features=[feature for feature in numerical_features if len(df[feature].unique())<25 and feature not in year_feature+['Id']]
```

```python
discrete_features
```

```python
#Let's find a relationship between  discrete features and and sales price
for feature in discrete_features:
    data=df.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
#List of continous features
continous_feature=[feature for feature in numerical_features if feature not in discrete_features + year_feature+['Id']  ]
```

```python
continous_feature
```

```python
for feature in continous_feature:
    data=df.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

**EXPLORATORY DATA ANALYSIS PART 2**


```python
#APPLYING LOGARITHMIC TRANSFORMATION FOR SKEWED DATA
for  feature in continous_feature:
  data = df.copy()
  if 0 in data[feature].unique():
    pass
  else:
    data[feature]=np.log(data[feature])
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

**OUTLIERS**(BOX PLOTS WORK ONLY FOR CONTINOUS VARIABLES)

```python
for feature in continous_feature:
    data=df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
        

```
